````python
#Importo paquetes necesarios
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
from scipy.integrate import odeint
import os
````

## El Atractor de Rössler X 10 ✵

El [sistema de Rössler](https://en.wikipedia.org/wiki/R%C3%B6ssler_attractor) es un sistema de ecuaciones diferenciales lineales que presenta comportamiento caótico. Las ecuaciones que definen al sistema son:

![](https://upload.wikimedia.org/math/7/b/8/7b8101ee4d46234d6ce0db24c1974a7a.png? raw=true =350x).

Siendo a, b y c parámetros.

###1. ✵✵ Resuelva las ecuaciones diferenciales para

```
a = b = 0.1,
c = 14,
```
y t en el intervalo [0,600.].

###Haga una gráfica de la trayectoria en 3D usando  `mpl_toolkits.mplot3d` y otra que sea una proyección sobre el plano `xy`.

````python
#Defino las ecuaciones del Sistema de Rössler.
def rossler(Y,t):
    a = 0.1
    b = 0.1
    c = 14
    x = Y[0]
    y = Y[1]
    z = Y[2]
    dx = -y -z
    dy = x + a*y
    dz = b + z*(x - c)
    return(np.array([dx,dy,dz]))
````

````python
#Defino el tiempo t y se da la solución a las ecuaciones diferenciales utilizanod odeint.
t = np.linspace(0,600,100000)
sol,others = odeint(rossler,[1.,1.,1.],t,full_output=True)
````

````python
#Hago la grafica en 3D con los datos arrojados por odeint.
x = sol[:,0]
y = sol[:,1]
z = sol[:,2]
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection='3d')
plt.plot(x,y,z)
````

````python
#Hago la gráfica en 2D en el plano xy.
fig = plt.figure(figsize=(12,12))
plt.plot(x,y)
````

###2. ✵✵ A continuación reproduzca el siguiete panel de gráficas (a su vez reproducido del artículo en Wikipedia sobre el sistema). En ello resuelva las ecuaciones diferenciales dejando `a=b=0.1` y resolviendo para el array de tiempos `np.linspace(0,600,10000)` eliminando los 2000 primeros puntos (el transiente).

````python
valores_c = np.array([4.,6.,8.5,8.7,9.,12.,12.6,13.,18.])
t = np.linspace(0,600,10000)

fig = plt.figure(figsize=(15,15))
plt.subplots_adjust(hspace = 0.2, wspace = 0.2)

for i in range(len(valores_c)):
    def nuevo_rossler(Y,t):
        a = 0.1
        b = 0.1
        c = valores_c[i]
        x = Y[0]
        y = Y[1]
        z = Y[2]
        dx = -y -z
        dy = x + a*y
        dz = b + z*(x - c)
        return(np.array([dx,dy,dz]))
    
    sol,others = odeint(nuevo_rossler,[1.,1.,1.],t,full_output=True)
    x = sol[:,0][2000:10000]
    y = sol[:,1][2000:10000]
    plt.subplot(3,3,int(i+1))
    plt.plot(x,y)
    plt.title('c=%d'%(valores_c[i]))
````

###3. ✵✵✵ Ahora construya el diagrama de bifurcación para una [sección de Poincaré](https://en.wikipedia.org/wiki/Poincar%C3%A9_map) aproximada por las condiciones: abs(x)<=1 and y >= 0. Tome `a=b=0.1` y modifique `c` entre 1 y 45. Al igual que en el anterior ejercicio elimine 2000 puntos para descartar el transiente.

````python
valores_c = np.linspace(1,45,90)
puntosc = np.array([])
puntosy = np.array([])
for i in range(len(valores_c)):
    def poincare_rossler(Y,t):
        a = 0.1
        b = 0.1
        c = valores_c[i]
        x = Y[0]
        y = Y[1]
        z = Y[2]
        dx = -y -z
        dy = x + a*y
        dz = b + z*(x - c)
        return(np.array([dx,dy,dz]))
     
    sol,others=odeint(poincare_rossler,[1.,1.,1.],t,full_output=True)
    
    x = sol[:,0][2000:10000]
    y = sol[:,1][2000:10000]
    
    for j in range(len(x)):
        if (np.absolute(x[j])<=1 and y[j] >= 0):
            puntosc = np.append(puntosc,valores_c[i])
            puntosy = np.append(puntosy,y[j])
````

````python
fig = plt.figure(figsize=(15,7))
plt.scatter(puntosc,puntosy,s=0.5)
plt.xlim(0,45)
plt.ylim(0,60)
plt.xlabel("c")
plt.title("Rössler system: Bifurcation Diagram\n a=b=0.1")
````

###4. ✵✵✵ Finalmente haga una animación que de cuenta de la estructura tridimiensional del atractor. **No** utilice `mpl_toolkits.mplot3d`, haga proyecciones sobre los siguientes planos:

  P(θ):={[x,y,z], n=[cos(45º) cos(θ), cos(45º) sin(θ), sin(45º) ] | [x,y,z].n = 0}

  Estos planos están generados por el siguiente par de vectores ortonormales:

  n1(θ) = [-sin(θ), cos(θ), 0]

  n2(θ) = [-cos(45º)cos(θ), -cos(45º)sin(θ), sin(45º)].

###El resultado debe ser similar a [este](https://github.com/ComputoCienciasUniandes/MetodosComputacionalesLaboratorio/raw/master/2015-V/actividades/talleres/Taller6/RosslerAnimation.mp4). Para ello tome los mismos parámetros que en el primer ejercicio duplicando el número de puntos y dejando solo los últimos 4000.

````python
#Creo una variable que contenga el doble de tiempo, utilizaré la función rossler definida en el punto (1)
#También resuelvo la ecuación diferencial
t2 = np.linspace(0,600,20000)
sol,others = odeint(rossler,[1.,1.,1.],t2,full_output=True)
````

````python
#Defino valores de x,y,z
x2 = sol[:,0][16000:20000]
y2 = sol[:,1][16000:20000]
z2 = sol[:,2][16000:20000]
````

````python
#Creo los vectores que definen el plano
theta = np.linspace(0,2*np.pi)

cos45 = -np.cos(np.pi/4)
sin45 = np.sin(np.pi/4)

n1 = np.array([[-np.sin(theta[0]),np.cos(theta[0]),0]])
n2 = np.array([[cos45*np.cos(theta[0]),cos45*np.sin(theta[0]),sin45]])

for i in range(1,len(theta)):
    n11 = np.array([-np.sin(theta[i]),np.cos(theta[i]),0])
    n22 = np.array([cos45*np.cos(theta[i]),cos45*np.sin(theta[i]),sin45])
    n1 = np.append(n1,[n11],axis=0)
    n2 = np.append(n2,[n22],axis=0)
````

````python
for i in range(len(n1)):
    puntos = []
    dotx = []
    doty = []
    for j in range(len(x2)):
        puntos.append([x2[j],y2[j],z2[j]])
        dotx.append(np.dot(puntos[j],n1[i]))
        doty.append(np.dot(puntos[j],n2[i]))
    fig = plt.figure(figsize=(30,15))
    plt.plot(dotx,doty)
    plt.xlim(-50,50)
    plt.ylim(-25,45)
    plt.axis('off')
    plt.savefig('anim%d.png'%(i))
    plt.cla()
    plt.close()
````

```python
#Importaci√≥n de paquetes
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.fftpack import fftshift as fts
from prettytable import PrettyTable
```

###(a)  Comenzando con un array de ceros de lado 1024 manipularlo con ciclos o usando *slice notation* para dejar todo en cero excepto elementos en un cuadrado de lado 32 en el centro.

```python
#Creo array cuadrado de 1024*1024 (inicialmente lleno de 0s)
placa = np.zeros(1024**2)
placa = np.resize(placa,(1024,1024))
```

```python
#Se comprueba que la placa est√© llena de zeros
plt.imshow(placa)
plt.colorbar()
```

```python
#Lleno en la mitad de la placa una secci√≥n de 32 de lado
placa[(1024-32)/2:(1024+32)/2,(1024-32)/2:(1024+32)/2] = 1
```

```python
#Se comprueba que la mitad de la placa NO est√© llena de zeros
plt.imshow(placa,cmap="Greys_r")
plt.colorbar()
```

###(b) Calcular la DFT del array resultante y aplicar sobre √©l la funci√≥n [fftshift](http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.fftpack.fftshift.html), esto hace un cambio de coordenadas que arroja el patr√≥n de interferencia esperado. De lo resultante tomar del centro un cuadrado de lado 200 y producir una imagen del logaritmo del m√≥dulo al cuadrado usando *imshow* con la representaci√≥n de la apertura a su izquierda.

```python
#Realizo la transformada de Fourier discreta
DFTplaca = np.absolute(np.fft.fft2(placa))
#Aplico scipy.fftpack.fftshift al array de la transformada
SDFTplaca = fts(DFTplaca)
```

```python
#Obtengo un cuadrado de lado 200 del centro de la placa
centroPlaca = SDFTplaca[(1024-200)/2:(1024+200)/2,(1024-200)/2:(1024+200)/2]
```

```python
#Obtengo el logar√≠tmo del m√≥dulo al cuadrado de los valores del centro de la placa
centroModificadoPlaca1 = np.log((centroPlaca)**2)
centroModificadoPlaca2 = np.log((centroPlaca+1)**2)
```

```python
#Se grafica este centro para comprobar
plt.figure(figsize=(12,8))
plt.subplots_adjust(hspace=0.00, wspace=0.00)

plt.subplot(2,2,1)
plt.imshow(centroModificadoPlaca1)
plt.colorbar()

plt.subplot(2,2,2)
plt.imshow(centroModificadoPlaca2)
plt.colorbar()

#Ac√° hago dos gr√°ficas. La primera es como aparece en el enunciado de la tarea, sin embargo me marca error en el
#logaritmo.
#En la segunda le agrego +1 para que no salga el error del logaritmo. Sin embargo, as√≠ no se ve como la del enunciado
```

###(d) Ahora h√°galo para una apertura circular de radio 0.01$\mu$ y adem√°s de lo anterior tome ahora la fila central y tomando como abscisas $\frac{2\pi r}{1.linspace(-512,512,1024)}$ reproduzca la gr√°fica mostrada abajo; las abscisas est√°n as√≠ elegidas para hacer comparable nuestro resultado con lo mostrado en la p√°g. 396 de los *Principles* de Born. $I_0$ es la amplitud m√°xima.

```python
#Apertura circular
aperturaCircular = np.zeros(1024**2)
aperturaCircular = np.resize(aperturaCircular,(1024,1024))

#Defino el radio
r=1024*0.01
```

```python
#Defino funcion del circulo
def circulo(x,y):
    return np.sqrt((x - 512)**2 + (y - 512)**2)
```

```python
#Relleno el c√≠rculo
for x in range(0,1024,1):
    for y in range(0,1024,1):
        if circulo(x,y) < r:
            aperturaCircular[x][y]= 1
```

```python
#Miro que el c√≠rculo est√© bien
plt.imshow(aperturaCircular,cmap="Greys_r")
```

```python
#Realizo la transformada de Fourier discreta
DFTaperturaCircular = np.absolute(np.fft.fft2(aperturaCircular))
#Aplico scipy.fftpack.fftshift al array de la transformada
SDFTcircular = fts(DFTaperturaCircular)
```

```python
#Obtengo el m√≥dulo al cuadrado de los valores del centro de freqs (sin resaltar con el log)
circularAperture = np.log((SDFTcircular+1)**2)
```

```python
#Se grafica este centro para comprobar
plt.imshow(circularAperture)
plt.colorbar()
```

```python
#Se toman los valores de i, i0 e i/i0
i = SDFTcircular[512]**2
i0 = SDFTcircular[512][512]**2
I = i/i0
```

```python
#Creo arrea de la fila central
abcisa = 2*np.pi*0.01/1.*np.linspace(-512,512,1024)
```

```python
#Creo la gr√°fica de comparaci√≥n de intensidades
plt.figure(figsize=(16,4))
plt.xlim(0,10)
plt.ylim(0,1)
plt.xlabel(r'$2\pi Rj/L$')
plt.ylabel(r'$I/I_0$')
plt.plot(abcisa,I)
plt.scatter(abcisa,I)
```

###(e) Usando las mismas ordenadas y abscisas del anterior literal calcular los m√°ximos y m√≠nimos entre 0. y 12.0. Al hacerlo debe utilizar interpolaci√≥n en alg√∫n momento. Compare sus resultados con la tabla mostrada en la p√°g. 397 de los *Principles*

```python
#Filtro la abcisa para los valores entre 0 y 12
aux = abcisa[abcisa>0]
nuevaAbcisa = aux[aux<12]
```

```python
#Filtro los nuevos valores de I que concuerden con la nueva abcisa
inicio = np.where(abcisa == nuevaAbcisa[0])[0]
fin = np.where(abcisa == nuevaAbcisa[len(nuevaAbcisa)-1])[0]
nuevaI = I[inicio:fin+1]
```

```python
#Creo la gr√°fica de comparaci√≥n de intensidades para abcisa entre 0 y 12
plt.figure(figsize=(16,4))
plt.xlim(0,12)
plt.ylim(0,1)
plt.xlabel(r'$2\pi Rj/L$')
plt.ylabel(r'$I/I_0$')
plt.plot(nuevaAbcisa,nuevaI)
plt.scatter(nuevaAbcisa,nuevaI)
```

```python
#Encuentro los m√≠nimos y m√°ximos locales, as√≠ como sus posiciones en el array de nuevaI

#Array de booleans, True indica m√°ximo o m√≠nimo
minimos = np.r_[True, nuevaI[1:] < nuevaI[:-1]] & np.r_[nuevaI[:-1] < nuevaI[1:], True]
maximos = np.r_[True, nuevaI[1:] > nuevaI[:-1]] & np.r_[nuevaI[:-1] > nuevaI[1:], True]

#Posiciones de m√°ximos y m√≠nimos en el array nuevaI
minLocation = np.where(minimos == True)[0]
maxLocation = np.where(maximos == True)[0]

#Valores num√©ricos de los m√°ximos y m√≠nimos
localMinsY = nuevaI[minLocation]
localMaxsY = nuevaI[maxLocation]

#Valores de abcisa de m√°ximos y m√≠nimos
localMinsX = nuevaAbcisa[minLocation]
localMaxsX = nuevaAbcisa[maxLocation]
```

```python
#Defino los valores te√≥ricos
maxTeoricos = np.array([1,0.0175,0.0042,0.0016])
minTeoricos = np.array([0,0,0,0])
```

```python
#Dado que hay 8 puntos de inflexi√≥n entonces hago una interpolaci√≥n polin√≥mica de grado 9
fit = np.polyfit(nuevaAbcisa,nuevaI,9)
regresion = np.poly1d(fit)
```

```python
#Grafico la regresi√≥n contra datos originales
x = np.linspace(0,12,191)

#Coeficiente de correlaci√≥n
cc = np.mean(np.corrcoef(nuevaI,regresion(x)))

plt.figure(figsize=(16,4))
plt.xlim(0,12)
plt.ylim(0,1)
plt.xlabel(r'$2\pi Rj/L$')
plt.ylabel(r'$I/I_0$')
plt.text(6,0.6,"Coeficiente de correlacion: %.20f"%cc)
plt.plot(x,regresion(x))
plt.scatter(nuevaAbcisa,nuevaI,color='red')
```

```python
#Encuentro los m√≠nimos y m√°ximos locales seg√∫n la regresi√≥n
Ireg = regresion(x)

#Array de booleans, True indica m√°ximo o m√≠nimo
minimosReg = np.r_[True, Ireg[1:] < Ireg[:-1]] & np.r_[Ireg[:-1] < Ireg[1:], True]
maximosReg = np.r_[True, Ireg[1:] > Ireg[:-1]] & np.r_[Ireg[:-1] > Ireg[1:], True]

#Posiciones de m√°ximos y m√≠nimos en el array Ireg
minLocationReg = np.where(minimosReg == True)[0]
maxLocationReg = np.where(maximosReg == True)[0]

#Valores num√©ricos de los m√°ximos y m√≠nimos
localMinsYreg = Ireg[minLocationReg]
localMaxsYreg = Ireg[maxLocationReg]
```
